## Different Data Sources

### `pandas_datareader.data.get_data_yahoo`

```python
from pandas_datareader.data import get_data_yahoo

code = '000001.SS'
start, end = '2017-4-1', '2018-4-1'
tt_ratio = 0.5
eo_ratio = 1.2

m1 = LPPL.Model(get_data_yahoo, code, start, end, tt_ratio, eo_ratio)
m1.plot()
```

### `tushare.get_hist_data`

```python
from tushare import get_hist_data

code = '600848'
start, end = '2017-4-1', '2018-4-1'
tt_ratio = 0.5
eo_ratio = 1.2

m2 = LPPL.Model(get_hist_data, code, start, end, tt_ratio, eo_ratio)
m2.plot()
```
